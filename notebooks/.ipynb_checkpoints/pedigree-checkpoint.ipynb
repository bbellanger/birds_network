```python
# Import the necessary python packages
import numpy as np
import pandas as pd
import scipy as sp
import matplotlib.pyplot as plt

# Import API
import urllib.request, json, csv

# Make a network and plot it
import networkx as nx
from pyvis import network as net
from pyvis.network import Network
import pygraphviz
```

```python
# Import the birds dataset
with urllib.request.urlopen("https://gracula.psyc.virginia.edu/birds/api/pedigree/?species=zebf") as url:
    data = pd.read_json(url)
# print(data)
```

```python
data["uuid"] = data["uuid"].str[:6]
data["name"] = data["name"].str[5:]
data["sire"] = data["sire"].str[5:]
data["dam"] = data["dam"].str[5:]
data
```

```python
# replace 'None' by 'origin'
data = data.replace({None: "Origin"})
data
```

```python
# Remove birds with no Dam or no Sire
# Define a function that drops rows
def filter_rows_by_values(df, col, values):
    return df[~df[col].isin(values)]

# data = filter_rows_by_values(data, "sire", [None])
# data
```

```python
# Use NetworkX to build the network out of the csv
# Define the variables to extract:
name = data["name"]
sire = data["sire"]
dam = data["dam"]
alive = data["alive"]
sex = data["sex"]
```

```python
# Make a network for the birds now!
G = nx.Graph(height="2080px", width="1850")

data = data.reset_index() # Make sure indexes pair with number of rows
for index, row in data.iterrows():
    G.add_nodes_from([(row['name'], {"alive": row['alive']})])
    G.add_edge(row['dam'], row['name'])
    G.add_edge(row['sire'], row['name'])

pos = nx.nx_agraph.graphviz_layout(G, prog="twopi", args="")
# pos = nx.kamada_kawai_layout(G)
nx.draw(G, node_color = "red", with_labels=True, font_weight='bold', pos = pos) # , pos = pos
plt.show()
```

```python
## Use Pyvis to generate a dynamical network representation
g = net.Network(height="750px", width="100%", font_color="white", select_menu=True, notebook=True) # bgcolor="#222222",
g = net.Network(directed = True, notebook=True)

node_degree = dict(G.degree) # Count the degree of the node

# Setting up node size attribute
nx.set_node_attributes(G, node_degree, 'size')

nxg = G
g.from_nx(nxg)
g.toggle_physics(True) # Toggle the physic in-between nodes
g.show(name="example.html", local=True, notebook=True)
# g.show('example.html')
```

```python
# The color or the form of the nodes aren't kept from NetworkX to Pyvis
# Let's build a network directly in Pyvis
F = Network(notebook=True, height="750px", width="100%")

# Add a fictive bird with 'None' attributes as origin
# With Pandas
F.add_node('Origin', 'Origin')

# set the physics layout for the network
F.barnes_hut(gravity=-80000, central_gravity=0.3, spring_length=250, spring_strength=0.001, damping=0.09, overlap=0)

for index, row in data.iterrows():  
    # Add nodes to the graph
    if row['alive'] == True:
        if row['sex'] == 'M':
            F.add_node(row['name'], row['name'], color='green', shape='box')
        elif row['sex'] == 'F':
            F.add_node(row['name'], row['name'], color='green', shape='circle')
        else:
            F.add_node(row['name'], row['name'], color='green', shape='diamond')
    else:
        if row['sex'] == 'M':
            F.add_node(row['name'], row['name'], title=['name'], color='red', shape='box')
        elif row['sex'] == 'F':
            F.add_node(row['name'], row['name'], title=['name'], color='red', shape='circle')
        else:
            F.add_node(row['name'], row['name'], title=['name'], color='red', shape='diamond')
    
    # F.add_edge(row['dam'], row['name'], color='blue')
    # F.add_edge(row['sire'], row['name'], color='blue')

F.show("example.html")
```
